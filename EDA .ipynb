```python
import pandas as pd
import numpy as np
import seaborn as sns
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
companies = ['AAPL' , 'MSFT'  , 'DELL' , 'GOOG']
end_time = datetime.datetime.now()
start_time = datetime.datetime(end_time.year-24, end_time.month, end_time.day)

for stock in companies:
    globals()[stock] = yf.download(stock, start_time, end_time)
    
company_list = [AAPL, GOOG, MSFT,  DELL]
company_name = ["APPLE", "GOOGLE", "MICROSOFT", "DELL"]

for company, com_name in zip(company_list, company_name):
    company["company_name"] = com_name
    
df = pd.concat(company_list, axis=0)
df.tail(10)
```

```python
import datetime
import calendar
 
def findDay(date):
    born = datetime.datetime.strptime(str(date), '%Y-%m-%d  %H:%M:%S').weekday()
    return (calendar.day_name[born])
name = []
for date in df['Date']:
    day = []
    dd = findDay(date)
    day.append(dd)
    name.append(day[0])
df['DOW'] = name
```

## Ploting Closing price for the companies

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Adj Close'].plot()
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {company_name[i - 1]}")
    
plt.tight_layout()
```

## Ploting Open prices for the companies

```python
plt.figure(figsize=(10, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Open'].plot()
    plt.ylabel('Open')
    plt.xlabel(None)
    plt.title(f"Open Prices of {company_name[i - 1]}")
    
plt.tight_layout()
```

```python
## Let us see how much stocks were trades for these companies
plt.figure(figsize=(10, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Volume'].plot()
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f"Stockes Traded of {company_name[i - 1]}")
    
plt.tight_layout()
```

```python
## Let us do futher analysis for the companies and see the daily risk and daily return of the stocks
for company in company_list:
    company['Daily Risk'] = company['Adj Close'].pct_change()
    
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(10)

AAPL['Daily Risk'].plot(ax=axes[0,0], legend=True, linestyle='--', marker='o')
axes[0,0].set_title('APPLE')

GOOG['Daily Risk'].plot(ax=axes[0,1], legend=True, linestyle='--', marker='o')
axes[0,1].set_title('GOOGLE')

MSFT['Daily Risk'].plot(ax=axes[1,0], legend=True, linestyle='--', marker='o')
axes[1,0].set_title('MICROSOFT')

DELL['Daily Risk'].plot(ax=axes[1,1], legend=True, linestyle='--', marker='o')
axes[1,1].set_title('DELL')

fig.tight_layout()
```
